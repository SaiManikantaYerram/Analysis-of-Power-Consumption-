Import the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Loading the data into a Data Frame by skipping few rows and labeling the columns 

refernece:https://www.youtube.com/watch?v=20eJss95TVo

```python
df1=pd.read_csv('RollingSystemDemand_20191012_1239.csv',
                         index_col=False,
                         skiprows=1,
                         skipfooter=1,
                         engine='python',
                         header=None)
df1.columns=['VD','Time','Value']
```

Converting the Time column data to proper Date Time format

reference:https://www.youtube.com/watch?v=20eJss95TVo

```python
df1
```

```python
df1['Time']=pd.to_datetime(df1['Time'].astype(str),format='%Y%m%d%H%M%S')
```

```python
df1.head()
```

```python
df1
```

Similarly reading the next part of data

```python
df2=pd.read_csv('RollingSystemDemand_20191012_1243.csv',
                         index_col=False,
                         skiprows=1,
                         skipfooter=1,
                         engine='python',
                         header=None)
df2.columns=['VD','Time','Value']
```

Changing Time column into the required format

```python
df2['Time']=pd.to_datetime(df2['Time'].astype(str),format='%Y%m%d%H%M%S')
```

```python
df2
```

combining the above two dataframes 

reference: https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

```python
final_df=pd.concat([df2,df1])
```

```python
final_df
```

Setting the index of the final dataset

```python
final_df.index= range(len(final_df))
```

Getting the hour from Time column and storing them in a new column

reference:https://stackoverflow.com/questions/25129144/pandas-return-hour-from-datetime-column-directly

```python
def hr_func(ts):
    return ts.hour

final_df['time_hour'] = final_df['Time'].apply(hr_func)
final_df['time_date'] = pd.DatetimeIndex(final_df['Time']).day
```

```python
final_df['time_hour']
```

```python
final_df['time_date']
```

Plotting the average/mean acroos all days for each hour

```python
fig_size = plt.rcParams["figure.figsize"]
print(fig_size)
fig_size[0] = 6
fig_size[1] = 4
plt.rcParams["figure.figsize"] = fig_size
final_df['Value'].groupby(final_df['time_hour']).mean().plot(kind='bar')
plt.xlabel('Hours')
plt.ylabel('Mean value(in MW)')
plt.title('Average/Mean acroos all days for each hour')
```

Mean power consumption for each hour on  each day

```python
final_df['Value'].groupby([final_df['time_date'],final_df['time_hour']]).describe()
```

Plotting the Mean power consumption for each hour on each day

The first plot was not that clear so plotted it horizontally using horizontal bar plot

horizontal bar hcart  reference:https://mode.com/example-gallery/python_horizontal_bar/
Axes on both sides of graph reference:https://stackoverflow.com/questions/6998697/in-matplotlib-how-do-you-display-an-axis-on-both-sides-of-the-figure

```python
fig_size = plt.rcParams["figure.figsize"]
print(fig_size)
fig_size[0] = 20
fig_size[1] = 100
plt.rcParams["figure.figsize"] = fig_size
final_df['Value'].groupby([final_df['time_date'],final_df['time_hour']]).mean().plot(kind='barh')
plt.tick_params(labeltop=True, labelright=True)
plt.ylabel('DATE,HOUR',fontsize=20)
plt.xlabel('Mean power consumption(in MW)',fontsize=20)
plt.title('Mean power consumption for each hour on each day',fontsize=20)
```
